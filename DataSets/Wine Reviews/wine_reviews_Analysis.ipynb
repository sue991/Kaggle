## Points

```python
mean_points = data.groupby(['country'])['points'].mean().to_frame().sort_values(by="points")
```

각 도시의 평균 point를 구하고 정렬해서 찍어본 것이다. 근데 각 도시마다의 와인 개수가 다르니 중요한 자료일지는 잘 모르겠다.

## Price

null값을 -1 으로 채워주기

```python
data['price'] = data['price'].fillna(-1)
```

```python
plt.figure(figsize=(10,8))
sns.boxplot('price',data=data)
```

3000달러짜리 와인이 진짜로 있는건지 아니면 잘못된 값인지 모르겠다.

1000달러 이상인 데이터는 없애기로 했다.

```python
data = data[data['price'] < 1000]
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot('points', 'price',data=data)
```

대체적으로 가격이 높을 수록 좋은 점수를 받는 경우가 많았다.

```python
mean_price = data.groupby(['country'])['price'].mean().sort_values(ascending=False).to_frame()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x = mean_price['price'], y = mean_price.index, color='purple')
```

도시 별 와인 평균 가격.   
이집트가 마이너스인 것은 NULL값이었던 가격을 -1로 해놨었기 때문

```python
max_price = data.groupby(['country'])['price'].max().sort_values(ascending=False).to_frame()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x = max_price['price'], y = max_price.index)
```

가장 비싼 와인이 있는 순서대로 나타낸 것.
